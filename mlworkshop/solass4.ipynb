```python
def mean(a):
    n=len(a)
    return sum(a)/n
def median(a):
    a=list(a)
    a.sort()
    n=len(a)
    if n%2==0:
        return (a[(n//2)-1]+a[(n//2)])/2 
    return a[n//2]
def std(a):
    m=mean(a)
    a=list(a)
    n=len(a)
    l=[((i-m)**2) for i in a]
    return (sum(l)/n)**0.5
def cov(a,b):
    a=list(a)
    b=list(b)
    n=len(a)
    l=[(i-mean(a))*(j-mean(b)) for i,j in zip(a,b)]
    return sum(l)/n
    

```

```python
import numpy as np

# Given dataset
data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]

# Step 1: Calculate the mean
mean = np.mean(data)

# Step 2: Calculate the standard deviation
std_dev = np.std(data)

# Step 3: Compute the z-scores
z_scores = [(x - mean) / std_dev for x in data]

# Step 4: Calculate the percentage of data within one and two standard deviations
within_one_std_dev = [z for z in z_scores if  z <= 1]
within_two_std_dev = [z for z in z_scores if  z <= 2]

percentage_within_one_std_dev = (len(within_one_std_dev) / len(data)) * 100
percentage_within_two_std_dev = (len(within_two_std_dev) / len(data)) * 100

# Display the results
print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Z-scores:", z_scores)
print(f"Percentage of data within one standard deviation: {percentage_within_one_std_dev:.2f}%")
print(f"Percentage of data within two standard deviations: {percentage_within_two_std_dev:.2f}%")
```

```python
import numpy as np

def rosenbrock(x, y):
    return (1 - x)**2 + (y - x**2)**2

def rosenbrock_gradient(x, y):
    df_dx = -2 * (1 - x) - 4 * x * (y - x**2)
    df_dy = 2 * (y - x**2)
    return np.array([df_dx, df_dy])

def gradient_descent(gradient_func, initial_point, learning_rate=0.001, tolerance=1e-6, max_iterations=10000):
    l=[]
    point = np.array(initial_point, dtype=float)
    for i in range(max_iterations):
        gradient = gradient_func(point[0], point[1])
        new_point = point - learning_rate * gradient
        if np.linalg.norm(new_point - point) < tolerance:
            break
        l.append(point)
        point = new_point
    plt.plot(l)
    plt.show()
    return point

# Initial solution
initial_point = (0, 0)

# Perform gradient descent
min_point = gradient_descent(rosenbrock_gradient, initial_point)

print(f"Minimum point: x = {min_point[0]}, y = {min_point[1]}")
print(f"Function value at minimum point: f(x, y) = {rosenbrock(min_point[0], min_point[1])}")
```
